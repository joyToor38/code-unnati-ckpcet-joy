# Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x axis value
x=[1,3,5,7,9]

# y axis value
y=[2,7,4,12,10]

# plotting the points
plt.plot(x,y)

# naming the x- axis // can name anything
plt.xlabel("sales")

# naming the y- axis
plt.ylabel("profit")

plt.title("First Graph")
```

```python
from numpy import random
i = random.randint(10, size = (5))
print("i :",i)
j = random.randint(10, size = (5))
print("j :",j)

plt.plot(i,j)
# naming the x- axis // can name anything
plt.xlabel("sales")

# naming the y- axis
plt.ylabel("profit")

plt.title("Graph made using random numbers")
```

```python
# Dotted graph

# x axis value
x=[1,3,5,7,9]

# y axis value
y=[2,7,4,12,10]

# plotting the points
plt.plot(x,y, linestyle = "dotted")

# naming the x- axis // can name anything
plt.xlabel("sales")

# naming the y- axis
plt.ylabel("profit")

plt.title("First Graph")
```

```python
# colouring graph

# x axis value
x=[1,3,5,7,9]

# y axis value
y=[2,7,4,12,10]

# plotting the points
plt.plot(x,y, color = "g")

# naming the x- axis // can name anything
plt.xlabel("sales")

# naming the y- axis
plt.ylabel("profit")

plt.title("First Graph")
```

```python
# Marking points

# x axis value
x=[1,3,5,7,9]

# y axis value
y=[2,7,4,12,10]

# plotting the points
plt.plot(x,y, color = "b", linestyle = "dotted", marker = "o")

# naming the x- axis // can name anything
plt.xlabel("sales")

# naming the y- axis
plt.ylabel("profit")

plt.title("First Graph")
```

```python
# Increasing marker size

# x axis value
x=[1,3,5,7,9]

# y axis value
y=[2,7,4,12,10]

# plotting the points
plt.plot(x,y, color = "b", marker = "o", ms = 15)

# naming the x- axis // can name anything
plt.xlabel("sales")

# naming the y- axis
plt.ylabel("profit")

plt.title("First Graph")
```

# Subplotting

```python
import numpy as np

#plot1
x=np.array([1,3,8,11])
y=np.array([3,8,1,10])

#plot2
a=np.array([5,7,9,10])
b=np.array([1,4,3,5])

plt.subplot(1,2,1) # row, column, location
plt.plot(x,y)

plt.subplot(1,2,2) 
plt.plot(a,b)
```

```python
# Changing location

plt.subplot(1,2,2) 
plt.plot(x,y)

plt.subplot(1,2,1) 
plt.plot(a,b)
```

```python
# up and down

plt.subplot(2,2,1) 
plt.plot(x,y)

plt.subplot(2,1,2) 
plt.plot(a,b)
```

```python
# up and down

plt.subplot(2,2,1) 
plt.plot(x,y)

plt.subplot(2,2,4) 
plt.plot(a,b)
```

```python
# plot 4 subgrpahs in one graph

c = np.array([1,3,5,7])
d = np.array([0,2,4,6])

e = np.array([2,3,5,9])
f = np.array([1,9,6,2])


plt.subplot(2,2,1) 
plt.plot(x,y)

plt.subplot(2,2,2) 
plt.plot(a,b)

plt.subplot(2,2,3) 
plt.plot(c,d)

plt.subplot(2,2,4) 
plt.plot(e,f)
```
